```python
import fiona
import requests
import json
import collections
import matplotlib as plt
import geopandas as gpd
import pandas as pd
import shapely
from shapely.ops import unary_union
#from shapely.geometry import shape, mapping
#from geojson import MultiPolygon

requests.adapters.DEFAULT_RETRIES = 1
```

```python
'''
Load the bmp data from Lin's Geopackage
'''
bmps = []

with fiona.open('scott-3e13e28703c24d779271dcd4beac8346.1614355867000.gpkg') as layer:
    for feature in layer:
        feature['bmp_geometry'] = feature.pop('geometry')
        bmps.append(feature)
```

```python
'''
A Function to format BMPs so the API will accept them.
'''

def format_bmp(bmp, count_type, count_geo):

    
    result = {}
    geo_type = ''

    try:
        result['bmp_geometry'] = bmp['bmp_geometry']
        result['bmp_type'] = bmp['properties']['practice_type']
        result['bmp_group'] = bmp['properties']['practice_type']
        result['model'] = 'gwlfe'
        result['id'] = bmp['id']
        geo_type = bmp['bmp_geometry']['type']
    except:
        pass
    count_type.append("BMP_type: " + bmp['properties']['practice_type'] + " Geo_Type: " + geo_type ) 
    return result
```

```python
'''
A function to send formated BMPs to the API.
'''

def model(data):
    results = requests.post('http://centrecounty.cci.drexel.edu/api/bmp/', json=data)
    return results
```

```python
def makePolygon(coordinate):
    polygon = {}
    polygon['type'] = 'LineString'
    polygon['coordinates'] = coordinate
    return polygon
```

```python
'''
Format BMPs
'''

count_type, count_geo = [], []
bmpfs = []
for bmp in bmps:
    bmpfs.append( format_bmp(bmp, count_type, count_geo))
count_type.sort()
bmp_cnt = collections.Counter(count_type)
#bmp_cnt
```

```python
for bmpf in bmpfs:
    if bmpf['id'] == '95':
        bmpf95 = bmpf
    if bmpf['id'] == '96':
        bmpf96 = bmpf      
#bmpf96
```

```python
'''
Start app update work

'''
geoms = shapely.geometry.shape(bmpf96['bmp_geometry'])
watersheds = []

if geoms.geom_type == 'Polygon':
    _x = json.dumps(shapely.geometry.mapping(geoms))
    _url = 'https://watersheds.cci.drexel.edu/api/watershedboundary/'
    _headers = {'Content-Type': 'application/json'}
    r = requests.post(_url, data = _x , headers= _headers, allow_redirects=True, verify=False)
    r = shapely.geometry.shape(r.json())
    watersheds.append(r)
elif geoms.geom_type == 'MultiPolygon':
    for geom in geoms:
        _x = json.dumps(shapely.geometry.mapping(geom))
        _url = 'https://watersheds.cci.drexel.edu/api/watershedboundary/'
        _headers = {'Content-Type': 'application/json'}
        r = requests.post(_url, data = _x , headers= _headers, allow_redirects=True, verify=False)
        r = shapely.geometry.shape(r.json())
        watersheds.append(r)
    
```
